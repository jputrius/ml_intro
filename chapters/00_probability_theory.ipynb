## Probability Space

A probability space is a triple $(\Omega, \mathcal{A}, \mathbb{P})$, where:

1. $\Omega$ is just some abstract set called the sample space.
2. $\mathcal{A}$ - the event space, which is a $\sigma$-algebra of subsets of $\Omega$. That is $\mathcal{A}$ is a set of some of the subsets of $\Omega$ such that:
    1. $\emptyset, \Omega \in \mathcal{A}$
    2. If $A \in \mathcal{A}$ then $\Omega \backslash A \in \mathcal{A}$
    3. If $(A_i)_{i=1}^\infty \subset \mathcal{A}$ then $\bigcup_{i=1}^\infty A_i, \bigcap_{i=1}^\infty A_i \in \mathcal{A}$

## Probability Space

3. $\mathbb{P}: \mathcal{A} \rightarrow [0, 1]$ is the probability measure, it must satisfy:
    1. If $(A_i)_{i=1}^\infty \subset \mathcal{A}$ are pairwise disjoint then $\mathbb{P}(\bigcup_{i=1}^\infty A_i) = \sum_{i=1}^\infty \mathbb{P}(A_i)$
    2. $\mathbb{P}(\Omega) = 1$

## Probability Space

This definition might seem rather abstract so let's look at an intuitive interpretation.

Suppose we were running some sort of a stochastic experiment, such as throwing a regular playing die. Then the sample space $\Omega$ represents all the possible outcomes of the experiment. In the playing die case we could take $\Omega = \{1, 2, 3, 4, 5, 6\}$.

## Probability Space

The event space $\mathcal{A}$ represents all the possible compound events that might occur, such as throwing a 3, throwing an even number etc. In our example we would take $\mathcal{A}$ to be all the subsets of $\Omega.$

The probability measure $\mathbb{P}$ is a (magical) device that measures the probability of events. In our concrete example,
$$
\mathbb{P}(\text{An odd number was thrown}) = \frac{1}{2}.
$$

## Independent Events

Two events $A, B \in \mathcal{A}$ are called independent if
$$
\mathbb{P}(A \text{ and } B) = \mathbb{P}(A) \mathbb{P}{B}
$$

For example if we were throwing a coin twice, then events "first throw lands heads" and "second throw lands tails" are independent, so
$$
\mathbb{P}(\text{first throw lands heads and second throw lands tails}) = \\
\mathbb{P}(\text{first throw lands heads}) \mathbb{P}(\text{second throw lands tails}) = \frac{1}{2} \frac{1}{2}.
$$

## Conditional probability

Suppose $A, B \in \mathcal{A}$ and $B$ has non-zero probability.
Then conditional probability is defined as
$$
\mathbb{P}(A | B) = \frac{\mathbb{P}(A \text{ and } B)}{\mathbb{P}(B)}.
$$
This is interpreted as the probability of $A$ happening if $B$ has already happened or we know will happen in the future.

## Bayes theorem

Suppose $A, B \in \mathcal{A}$ and $A$ and $B$ both have non-zero probability.
Then we have the Bayes formula
$$
\mathbb{P}(A | B) = \frac{\mathbb{P}(B | A)\mathbb{P}(A)}{\mathbb{P}(B)}.
$$

Also,
$$
\mathbb{P}(A \text{ and } B) = \mathbb{P}(A | B)\mathbb{P}(B).
$$

## Random variables

A random variable is a measurable function $X: \Omega \rightarrow \mathbb{R}$, that is for any interval $(a, b)$ we have $X^{-1}((a, b)) \in \mathcal{A}.$

For example, if we were throwing two dice, then the sum of the throws would be a random variable.

## Discrete random variables

A random variable $X$ is called discrete if it takes only finite or countable number of values.
For discrete random variables the point mass function (pmf) is defined
$$
p_X(x) = \mathbb{P}(X=x).
$$
It is non-zero for at most a countable number of values.

## Mean

The mean or expectation of a discrete random variable is defined to be
$$
\mathbb{E}[X] = \sum_x x p_X(x),
$$
where we sum over those $x'$s, where $p_X(x) > 0.$ The brackets [] don't indicate anything special, its just tradition to use them.

The expectation is linear, that is if $X$ and $Y$ are random variables and $a, b \in \mathbb{R}$ then
$$
\mathbb{E}[aX+bY] = a\mathbb{E}[X] + b \mathbb{E}[Y].
$$

## Variance

The variance of the random variable $X$ is defined to be
$$
\text{Var}(X) = \mathbb{E}[(X - \mathbb{E}[X])^2] = \mathbb{E}[X^2] - \mathbb{E}[X]^2.
$$
The standard deviation is the square root of variance
$$
\sigma = \sqrt{\text{Var}(X)}.
$$

## Variance

Variance is not linear. If $X$ and $Y$ are random variables and $a, b \in \mathbb{R}$ then
$$
\text{Var}(aX+bY) = a^2\text{Var}(X)+b^2\text{Var}(Y) + 2ab \text{Cov}(X, Y),
$$
where 
$$
\text{Cov}(X, Y) = \mathbb{E}[(X - \mathbb{E}[X])(Y - \mathbb{E}[Y])].
$$

## Law of Large Numbers

We will use random variables to model datapoints. For example we have some observations $x_1, \dots, x_n$ of, let's say, heights of penguins. Then we will imagine that these numbers $x_1, \dots, x_n$ are separate instantiations of a random variable $X,$ which represents the heights of penguins.

If $\mathbb{E}[X]$ is finite then the law of large numbers holds, that is
$$
  \frac{\sum_{i=1}^n x_n}{n} \approx \mathbb{E}[X].
$$

## Law of Large Numbers

How good the approximation is depends on $\text{Var}(X)$. If $\text{Var}(X)$ is infinite then the approximation is really really bad and you shouldn't use law of large numbers in practice in such a case (but people still do).

## Correlation

Suppose $X$ and $Y$ are random variables, then their correlation coefficient is defined to be
$$
  \text{corr}(X, Y) = \frac{\text{Cov}(X, Y)}{\sqrt{\text{Var}(X)}\sqrt{\text{Var}(Y)}}.
$$
It is a number between $-1$ and $1$ (if it exists).

If the correlation coefficient is 0 then the random variables are said to be uncorrelated or linearly independent.

## Correlation

To help interpret correlation let's look at the following example. Suppose we have two uncorrelated random variables $X$ and $Y$.
Now take numbers $a, b \in \mathbb{R}$ such that $a^2+b^2=1$ and $aX+bY$ would have the smallest possible variance out of all choices of $a$ and $b$ with the specified condition.
Then if $\text{Var}(X) \le \text{Var}(Y)$, we have $a=1$ (or $a=-1$) and $b=0$.

That is, if two variables are uncorrelated then there does not exist a linear model between the variables that would be at all viable.

## Correlation

Caution: Being uncorrelated does not imply that the variables are independent. For example $X$ and $Y$ might be uncorrelated, but $X^2$ and $Y$ can possibly be strongly correlated.

## Continuous Random Variables

A random variable $X$ is called absolutely continuous if there exists a function $f_X: \mathbb{R} \rightarrow \mathbb{R}$ such that for any two real numbers $a < b$ we can compute the probability as
$$
\mathbb{P}(a \le X \le b) = \int_a^b f_X(x) dx.
$$
The function $f_X$ is then called the point density function (pdf) of $X$.

## Continuous Random Variables

The mean for absolutely continuous random variables is defined as
$$
\mathbb{E}[X] = \int_{-\infty}^{\infty} x f_X(x) dx.
$$
